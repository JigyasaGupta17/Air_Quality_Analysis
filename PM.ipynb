```python
import pandas as pd
import os


def calculate_monthly_means(folder_path, pm_columns=["PM2.5 (ug/m3)", "PM10 (ug/m3)"]):
    """
    Calculates monthly mean PM2.5 and PM10 concentrations from CSV files within a folder,
    ignoring files without the specified columns and printing the results.

    Args:
        folder_path (str): Path to the folder containing CSV files.
        pm_columns (list, optional): List of column names containing PM2.5 and PM10 concentrations.
            Defaults to ["PM2.5 (ug/m3)", "PM10 (ug/m3)"].

    Returns:
        None
    """

    missing_files = []

    # Create an empty DataFrame to store combined monthly means
    combined_means = pd.DataFrame(columns=["Month", "Year", "PM2.5_mean", "PM10_mean"])

    # Iterate through all CSV files in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith(".csv"):
            file_path = os.path.join(folder_path, filename)

            try:
                # Read the CSV file
                df = pd.read_csv(file_path)

                # Check for missing columns
                missing_cols = [col for col in pm_columns if col not in df.columns]
                if missing_cols:
                    missing_files.append(filename)
                    print(f"Skipping file '{filename}' due to missing columns: {', '.join(missing_cols)}")
                    continue

                # Extract month and year from the datetime column (handle potential errors)
                try:
                    if "To Date" in df.columns:  # Check if "To Date" column exists
                        df["month"] = pd.to_datetime(df["To Date"]).dt.month_name()
                        df["year"] = pd.to_datetime(df["To Date"]).dt.year
                    else:
                        print(f"Column 'To Date' not found in file: {filename}")
                        continue
                except KeyError:
                    print(f"Error extracting month and year from column in file: {filename}")
                    continue

                # Calculate monthly means
                monthly_means = df.groupby(["month", "year"])[pm_columns].mean().reset_index()

                # Print the monthly means for this file
                print(f"\nMonthly means from file '{filename}':")
                print(monthly_means)

            except FileNotFoundError:
                print(f"File not found: {file_path}")

    # Print information about files with missing columns
    if missing_files:
        print(f"\nFiles skipped due to missing columns:")
        for filename in missing_files:
            print(f"- {filename}")

    print(f"\nMonthly mean PM2.5 and PM10 concentrations calculated and printed.")
    user_choice = input("Do you want to open the folder containing the data files? (y/n): ")
    if user_choice.lower() == 'y':
        print(f"The folder containing the data files is located at: {folder_path}")
        print("Please use your file explorer to navigate to this location.")

# Example usage
folder_path = r"C:\Users\welcome\Pictures\Camera Roll\air quality state wise"
calculate_monthly_means(folder_path)
```
